```python
import requests
import urllib
from bs4 import BeautifulSoup

query = input("請使用者打欲查詢之關鍵字：")
url = "https://www.google.com.tw/search?hl=ch&q=" + urllib.parse.quote(query)
url_list = []

for i in range (3):
    res = requests.get(url , timeout = 3) 
    soup = BeautifulSoup(res.text,"html.parser")
    print ("第"+str(i+1)+"頁的URL為:\t"+ url ); url_list.append(url)
    
    next_link = soup.find(class_= 'nBDE1b G5eFlf').get('href') 

    url = "https://www.google.com/"+ next_link 
```

```python
q = 0 
for urlist in url_list:
    print("第 %d 頁搜尋結果的每個超連結為:\n" %(q+1));q+=1
    
    req = urllib.request.Request(url=urlist, headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:23.0) Gecko/20100101 Firefox/23.0'})
    soup = BeautifulSoup(urllib.request.urlopen(req).read(), features='lxml')
    
    tab_s= soup.find_all("div", class_="r")
    for tab in tab_s :
        print('%s\t%s'%(tab.a['href'],tab.find("h3", class_="LC20lb DKV0Md").text))
        
    print("-----------------------------------------------------------------")
```

```python
#舊版爬蟲

from bs4 import BeautifulSoup
from urllib.request import urlopen
import re

path = "C:/Users/jeffh/Desktop/EEG/TT/all2.txt"
file= open(path,'w',encoding='utf8')

for i in range(42649,42650): #42649 42664
    
    html = urlopen("https://www.healthnews.com.tw/news/article/"+"%s"%i+"/").read().decode('utf-8',errors='ignore')
    file.write('%s'%i)
    if len(html) != 0:
        html = " ".join(html.split("<br/>"))      
        soup = BeautifulSoup(html, features='lxml')
        all_hrefs = soup.find_all('p')

        #抓標題
        a_tags=soup.find_all('title')
        for tag in a_tags:
            print(tag.string,"    ", end="")
            file.write('%s'%tag.string)
            file.write('\n')
        #print("   主題:",end="")
        print("\n")
        #抓內容
        all_hrefs = soup.find_all('div', {'class': re.compile("caption")}) #id="newsContent"
        for all_href in all_hrefs:
            print(all_href.string," ") #換行 不換行, end=""
            file.write('%s'%all_href.string)
            file.write('\n')

    print("\n")#這張已經完成 故下一列
file.close
```
